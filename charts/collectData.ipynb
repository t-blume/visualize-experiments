```python
def parse_app_name(app_name, analyze_app):
    if app_name.startswith(analyze_app):
        version = app_name.replace(analyze_app, '')
        if version:
            if '_batch_' in version:
                iteration_number = int(version.replace('_batch_', ''))
                return analyze_app, iteration_number, True
            else:
                iteration_number = int(version)
                return analyze_app, iteration_number, False
    else:
        return None, None, None
```

```python
# debug
events = set()

df = pd.DataFrame(columns=['Iteration', 'Memory Incremental', 'Memory Batch'])

search_for_apps = 'dyldo_y2019_core_type-test-1'
for file in tqdm(onlyfiles):
    app_name = None
    peak_memory = 0
    max_memory = 0
    with open(file) as f:
        content = f.readlines()
        for line in content:
            event = json.loads(line)
            events.add(event['Event'])
            if event['Event'] == 'SparkListenerEnvironmentUpdate':            
                app_name = event['Spark Properties']['spark.app.name']
                app_name, iteration, batch_comp = parse_app_name(app_name, search_for_apps)
                if app_name:
                    if df.loc[df['Iteration'] == iteration].empty:
                        df.loc[len(df)] = iteration
            if event['Event'] == 'SparkListenerTaskEnd':
                for task in event['Task Info']['Accumulables']:
                    if task['Name'] == 'internal.metrics.peakExecutionMemory':
                        # peak_memory = max(task['Value'], peak_memory)
                        peak_memory += task['Value']
            if event['Event'] == 'SparkListenerBlockManagerAdded':
                # print(event['Maximum Memory'])
                max_memory = max(event['Maximum Memory'], max_memory)
        
        if app_name:
            # the droids we are looking for
            if batch_comp:
                df.loc[df['Iteration'] == iteration, 'Memory Batch'] = peak_memory
            else:
                df.loc[df['Iteration'] == iteration, 'Memory Incremental'] = peak_memory

    #display(events)
    # print('%s: peak memory at %.2f MB' % (app_name, peak_memory/1024/1024))
    #print('%s: max memory at %.2f MB' % (app_name, max_memory/1024/1024))
    
df = df.set_index('Iteration')
sorted_df = df.sort_values(by='Iteration')
```
